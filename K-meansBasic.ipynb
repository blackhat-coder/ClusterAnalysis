```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=300, n_features=2, centers=4, cluster_std=0.8, random_state=0)

X = data[0][: , 0]
Y = data[0][: , 1]
c = data[1]
```

```python
new_data = pd.DataFrame({'x':X , 'y':Y , 'centroids':c})
new_data.head(5)
```

```python
plt.figure(figsize=(10, 9))
sns.set_style('whitegrid')
sns.scatterplot(x=new_data['x'] , y=new_data['y'], c=new_data['centroids'], size=new_data['centroids'])
```

```python
# new_data.drop('centroids', inplace=True, axis=1)
```

```python
new_data.shape
```

```python
from sklearn.cluster import KMeans

k_model = KMeans(n_clusters=4)
y_pred = k_model.fit_predict(new_data)
```

```python
y_pred
```

```python
k_model.cluster_centers_
```

# KMeans

```python
new_data['k_labels'] = y_pred

plt.figure(figsize=(10, 9))
sns.set_style('whitegrid')
sns.set_color_codes(palette='deep')

plt.title('Kmeans')
sns.scatterplot(x=new_data['x'], y=new_data['y'], c=new_data['k_labels'], size=new_data['k_labels'])
plt.scatter(k_model.cluster_centers_[:, 0] , k_model.cluster_centers_[:, 1], s=100)
```

The Blue points on the scatterplot show the points that Kmeans converged on as centroids
