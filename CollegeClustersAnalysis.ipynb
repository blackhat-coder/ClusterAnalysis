```python


def rename_col(col_name, string_r):
    data.rename(columns={col_name: string_r}, inplace=True)

    print(f'renaming:{col_name} -> : {string_r}')
    print('')

rename_col('Apps', 'applications')
rename_col('Accept', 'accepted')
rename_col('Enroll', 'enrolled')
rename_col('F.Undergrad', 'full_time_undergrad')
rename_col('P.Undergrad', 'partime_undergrad')
rename_col('Room.Board', 'Room & Board costs')
rename_col('Books', 'books_cost')
rename_col('Personal', 'personal_spending')
rename_col('PhD', 'PhD_faculties')
rename_col('Terminal', 'Terminal_faculties')
rename_col('S.F.Ratio', 'student_faculty_ratio')
rename_col('perc.alumni', 'alumni_donations_pct')
rename_col('Expend', 'Instructional_expenditure')
rename_col('Grad.Rate' , 'grad_rate')
rename_col('Unnamed: 0', 'university_name')
```

```python
data.drop(['university_name', 'target'], axis=1, inplace=True)
```

```python
from sklearn.decomposition import PCA

pca_model = PCA(n_components=4)

compressed_data = pca_model.fit_transform(data)

compressed_data.shape
```

```python


from sklearn.cluster import KMeans

kmodel = KMeans(n_clusters=2)

y_pred = kmodel.fit_predict(data)
yc_pred = kmodel.fit_predict(compressed_data)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

a_score = accuracy_score(target, y_pred)
f1 = f1_score(target, y_pred)
c_matrix = confusion_matrix(target, y_pred)


ac_score = accuracy_score(target, yc_pred)
fc1 = f1_score(target, yc_pred)
cc_matrix = confusion_matrix(target, yc_pred)
```

```python
print(f' accuracy score: {a_score}')
print(f' f1_score: {f1}')

print(' ')

print('CONFUSSION MATRIX')
print(c_matrix)
```

```python
print(f' accuracy score: {ac_score}')
print(f' f1_score: {fc1}')

print(' ')

print('CONFUSSION MATRIX')
print(cc_matrix)
```
